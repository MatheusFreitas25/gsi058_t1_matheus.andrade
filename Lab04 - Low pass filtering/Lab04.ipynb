1)	Aplique o filtro de Média a imagem abaixo para diferentes tamanhos de máscaras.
- 3x3; 7x7; 3x3 aplicado três vezes sequenciais.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_original = cv2.imread('sta2.png', cv2.IMREAD_GRAYSCALE)
img_seq_1 = cv2.blur(img_original, (3, 3))
img_seq_2 = cv2.blur(img_seq_1, (7, 7))
img_seq_3 = cv2.blur(img_seq_2, (3, 3))

plt.figure(1)
plt.imshow(img_original, cmap='gray')
plt.show()

plt.figure(2)
plt.imshow(img_seq_1, cmap='gray')
plt.show()

plt.figure(3)
plt.imshow(img_seq_2, cmap='gray')
plt.show()

plt.figure(4)
plt.imshow(img_seq_3, cmap='gray')
plt.show()

img1 = cv2.imread("sta2.png", cv2.IMREAD_GRAYSCALE)
```

2)	Aplique o filtro Gaussiano à imagem abaixo para diferentes valores de desvio padrão (σ) e tamanho de máscara.
- σ = 1.0 (kernel size 5×5).	σ = 2.0 (kernel size 9×9).	σ =  4.0 (kernel size 15×15).

```python
#  code here
img_original = cv2.imread("ben2.png", cv2.IMREAD_GRAYSCALE)

img_filtro_gaussiano = cv2.GaussianBlur(img_original, (5, 5), 1.0)

plt.figure(1)
plt.imshow(img_filtro_gaussiano, cmap='gray')
plt.show()

img_filtro_gaussiano = cv2.GaussianBlur(img_original, (9, 9), 2.0)

plt.figure(2)
plt.imshow(img_filtro_gaussiano, cmap='gray')
plt.show()

img_filtro_gaussiano = cv2.GaussianBlur(img_original, (15, 15), 4.0)

plt.figure(3)
plt.imshow(img_filtro_gaussiano, cmap='gray')
plt.show()
```
